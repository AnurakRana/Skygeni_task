```python
df2=df2[df2['renewed']==True]
# take only the true renewed ones here for now
```

```python
# create a column that has the inflation rate of that specific time using the df basic plan is that the 
# date below the subcription get assign min of the inlfation
# and the end date that have beyond the end_date of our df gets assigned the max 
def func(date):
    row=df[(df['start_date']<=date)&(df['end_date']>=date)]
    if not row.empty:
        return row['inflation_rate'].values[0]
    elif df['start_date'].min()>date:
        return df['inflation_rate'].min()
    elif df['end_date'].max()<date:
        return df['inflation_rate'].max()
df2['inflation']=df2['end_date'].apply(func)
```
